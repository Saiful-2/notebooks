```python
# ============================================================
# Step 1: Import Required Libraries
# ============================================================

import pandas as pd
import numpy as np

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
# ============================================================
# Step 2: Load Dataset
# ============================================================

url = "https://raw.githubusercontent.com/Saiful-2/datasets/main/spam_ham_dataset.csv"

df = pd.read_csv(url)

print("Dataset Loaded Successfully")
df.head()
```

```python
# ============================================================
# Step 3: Explore Dataset
# ============================================================

print("\nDataset Shape:")
print(df.shape)

print("\nColumn Names:")
print(df.columns)

print("\nMissing Values:")
print(df.isnull().sum())
```

```python
# ============================================================
# Step 4: Select Input and Output Columns
# ============================================================

# Usually the dataset contains:
# text column = message/email text
# label column = spam/ham

X = df['text']
y = df['label']

print("\nSample Email:")
print(X.iloc[0])

print("\nSample Label:")
print(y.iloc[0])
```

```python
# ============================================================
# Step 5: Convert Text into Numerical Form
# Using TF-IDF Vectorization
# ============================================================

vectorizer = TfidfVectorizer(stop_words='english')

X_features = vectorizer.fit_transform(X)

print("\nText Converted into Numerical Features")
print(X_features.shape)
```

```python
# ============================================================
# Step 6: Split Dataset into Training and Testing Data
# ============================================================

X_train, X_test, y_train, y_test = train_test_split(
    X_features,
    y,
    test_size=0.2,
    random_state=42
)

print("Training Data Shape:", X_train.shape)
print("Testing Data Shape:", X_test.shape)
```

```python
# ============================================================
# Step 7: Apply Logistic Regression
# ============================================================

model = LogisticRegression()

model.fit(X_train, y_train)

print("Model Trained Successfully")
```

```python
# ============================================================
# Step 8: Make Predictions
# ============================================================

y_pred = model.predict(X_test)
```

```python
# ============================================================
# Step 9: Evaluate Accuracy
# ============================================================

accuracy = accuracy_score(y_test, y_pred)

print("\nAccuracy Score:")
print(round(accuracy * 100, 2), "%")
```

```python
# ============================================================
# Step 10: Display Sample Predictions
# ============================================================

print("\nSample Predictions:\n")

sample_texts = [
    "Congratulations! You have won a free iPhone. Click here now.",
    "Please attend the meeting at 10 AM tomorrow.",
    "You have been selected for a cash prize claim now.",
    "Project report submission deadline is tomorrow."
]

sample_features = vectorizer.transform(sample_texts)

predictions = model.predict(sample_features)

for text, pred in zip(sample_texts, predictions):
    print("Email:", text)
    print("Prediction:", pred)
    print("-" * 50)
```

```python
# ============================================================
# Step 11: Predict a Custom Email
# ============================================================

new_email = ["Win a lottery worth $5000 now!"]

new_email_features = vectorizer.transform(new_email)

prediction = model.predict(new_email_features)

print("\nCustom Email Prediction:")
print("Email:", new_email[0])
print("Prediction:", prediction[0])
```
